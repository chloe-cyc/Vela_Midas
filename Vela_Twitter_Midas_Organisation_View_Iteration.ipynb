```python
#Import Packages
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import datetime #bi

pd.options.mode.chained_assignment = None
```

```python
likes = pd.read_csv("Twitter Likes.csv")
likes = likes.drop_duplicates()
#midas["round_order"] = midas.round_name.map(round_order)
likes.head()
```

```python
likes_organization = likes.drop(columns=["investor_name","investor"]).groupby(likes["liked_user"], as_index=False).agg({"organization":"first","liked_user":"first", "count":"sum"})
```

```python
#List of investors
master_investor_list_likes = likes.investor.unique().tolist()
#List of orginizations
master_organization_list_likes = likes_organization.organization.unique().tolist()
```

```python
#ORGANIZATION DATA
organization_data = likes.drop(columns=["investor_name","investor"]).groupby(likes["liked_user"], as_index=False).agg({"organization":"first","liked_user":"first", "count":"sum"})
organization_data.head(300)
```

```python
"""
Return for a pair of investors or organizations, the list of common liked or commented users

Parameters
---
investor_1: str
  Username of the first investor that the function analyses or name of the organization
investor_2: str
  Username of the second investor that the function analyses or name of the organization
invororg1: str
  Whether the investor of interest is an organizationor an investor

Other things to consider that could be added as a parameter:
  Is the common user also an investor?
  Is the common user someone who has been interacted greater than x number of times?
"""

def liked_together(investor_1, investor_2, invororg1, invororg2):
  data = likes

  #list of names that investor_1 and investor_2 have interacted with individually
  if invororg1 == "investor":
    investor1_liked = data.loc[data[invororg1] == investor_1].liked_user.unique()
  elif invororg1 == "organization":
    investor1_liked = organization_data.loc[organization_data[invororg1] == investor_1].liked_user.unique()

  if invororg2 == "investor":
    investor2_liked = data.loc[data[invororg2] == investor_2].liked_user.unique()
  elif invororg2 == "organization":
    investor2_liked = organization_data.loc[organization_data[invororg2] == investor_2].liked_user.unique()     
  
  #List of users who have been interacted with by the two investors
  matchlikedusers = set(investor1_liked) & set(investor2_liked)
  print(matchlikedusers)
  return matchlikedusers

def not_liked_together(investor_1, investor_2, invororg1, invororg2):
  data = likes

  #list of names that investor_1 and investor_2 have interacted with individually
  if invororg1 == "investor":
    investor1_liked = data.loc[data[invororg1] == investor_1].liked_user.unique()
  elif invororg1 == "organization":
    investor1_liked = organization_data.loc[organization_data[invororg1] == investor_1].liked_user.unique()

  if invororg2 == "investor":
    investor2_liked = data.loc[data[invororg2] == investor_2].liked_user.unique()
  elif invororg2 == "organization":
    investor2_liked = organization_data.loc[organization_data[invororg2] == investor_2].liked_user.unique()

  otherlikedusers = set(investor1_liked).difference(investor2_liked)

  return otherlikedusers

```

```python
#For a given two investors who have both invested in a given user, the distance between them is given by subtracting the count of each
#n.b This function will throw an error if the user is not liked by both investors - which is fine in our case

def distance(investor_1, investor_2, user, invororg1 = "", invororg2 = ""):
  data = likes

  if invororg1 == "investor":
    count_1 = data[(data[invororg1]== investor_1) & (data['liked_user']== user)]["count"].item()
  elif invororg1 == "organization":
    count_1 = organization_data[(organization_data[invororg1] == investor_1) & (data['liked_user']== user)]["count"].item()

  if invororg2 == "investor":
    count_2 = data[(data[invororg2]== investor_2) & (data['liked_user']== user)]["count"].item()
  elif invororg2 == "organization":
    count_2 = organization_data[(organization_data[invororg2] == investor_2) & (data['liked_user']== user)]["count"].item()

  dist = abs(count_1-count_2)
  return dist

distance("ryan_mcintyre","JayRughani","pmarca", "investor", "investor")  
```

```python
"""
Finds N most similar investors to user specified constraints.

Parameters
---
Investor: Str
  The name of the Investor for which neighbours are which to be computed
N: Int
  The number of nearest neighbours to be computed
Similarity: str or list
  If str:
    Function used to compute the similarity between investors.
        - "jacc": jaccard similarity
        - "cond": conditional probability-based similarity
        - "abs": total number of users co-interacted
        - "count_sim": similarity which is inversly proportional to the distance between the investors specified using 'count_scale' and note that this 
Min_no_likes: int
  Only consider users which have at least minimum number of likes by specific user or total - not sure yet which one to use
sim_func: function
  The function used to convert the distance into a similarity measure on a [0, 1] interval when using like similarity as a similarity measure.
investors_list: list or ()
  If (): consider all the investors in the dataset
        Otherwise: only consider the similarity between the VC and the investors in the list
similarity_type: str
  - Investor to investor: "i_to_i"
  - Investor to organisation:"i_to_o"
  - Organisation to Organisation: "o_to_o" 
  - Organisation to investor: "o_to_i"

"""

def n_nearest_neighbours(investor_name, n=10, similarity="jacc", min_no_likes=50, sim_func="default", similarity_type = "i_to_i"):
  data = likes

  if sim_func == "default":
    def sim_func(x):
        return(1/(1+x))

  if similarity_type == "i_to_o":
    invororg1 = "investor"
    invororg2 = "organization"
    investor_un = data.loc[data.investor_name == investor_name, "investor"].head(1).item()
    investor_array = np.array(organization_data.groupby(invororg2).filter(lambda x: len(x) >= min_no_likes).organization.unique())
  elif similarity_type == "o_to_i":
    invororg1 = "organization"
    invororg2 = "investor"
    investor_un = investor_name
    investor_array = np.array(organization_data.groupby(invororg2).filter(lambda x: len(x) >= min_no_likes).organization.unique())
    investor_array_names = [data.loc[data.investor == i, "investor_name"].head(1).item() for i in investor_array]

  elif similarity_type == "i_to_i":
    invororg1 = "investor"
    invororg2 = "investor"
    investor_un = data.loc[data.investor_name == investor_name, "investor"].head(1).item()
    investor_array = investor_array[investor_array != investor_un] #List of investor usernames or organizations that do not include investor of interest
    investor_array_names = [data.loc[data.investor == i, "investor_name"].head(1).item() for i in investor_array]

  elif similarity_type == "o_to_o":
    invororg1 = "organization"
    invororg2 = "organization"
    investor_un = investor_name
    investor_array = np.array(organization_data.groupby(invororg2).filter(lambda x: len(x) >= min_no_likes).organization.unique())
    investor_array = investor_array[investor_array != investor_un] #List of investor usernames or organizations that do not include investor of interest

  #Computng similarities
  investor_interactions =[]
  for i in range(len(investor_array)):
    investor_interactions.append(data.loc[data[invororg2] == investor_array[i],"count"].sum())

  if invororg2 == "organization":
    similarity_df = pd.DataFrame({"Count": investor_interactions}, index=investor_array)
  elif invororg2 == "inveestor":
    similarity_df = pd.DataFrame({"Count": investor_interactions}, index=investor_array_names)
                      

  ind = 1
  if type(similarity) == str:  
    similarity = [similarity]

  for sim in similarity:

    if sim == "jacc": #counts have not been factored into this.
      sim_vector = [len(liked_together(investor_un,i,invororg1,invororg2))/(len(not_liked_together(investor_un,i,invororg1,invororg2))+len(liked_together(investor_un,i,invororg1,invororg2))) for i in investor_array]

    elif sim == "cond": #COunts have not been factored into this
      #Conditional probability that investor has liked a given user given that another investor has already liked them
      sim_vector = [len(liked_together(investor_un,i,invororg1,invororg2))/len(data.loc[data[invororg2] == i].liked_user.unique()) for i in investor_array] 

    elif sim == "abs": #counts have not been factored into this
      #Count the number of uesers that investor A and investor B have liked together
      sim_vector = [len(liked_together(investor_un,i,invororg1,invororg2)) for i in investor_array]

    elif sim == "count_sim":
      "Two investors are more similar if they liked a user a similar nmber of times."
      sim_vector = np.zeros(len(investor_array))
      dist = []
      total_dist = 0
      for i in range(len(investor_array)):
        together = liked_together(investor_un, investor_array[i],invororg1,invororg2)
        if together != 0:
          for u in together: # for a given user in the list of users that have been liked by two investors
            dist.append(distance(investor_un, investor_array[i], u,invororg1,invororg2))
            ave_dist = sum(dist)/len(together)
          sim_vector[i] = sim_func(ave_dist)
        elif together ==0:
          sim_vector[i]=0

    if invororg2 == "organization":
      new_sim_df = pd.DataFrame({sim: sim_vector}, index=investor_array)  
    elif invororg2 == "investor":
      new_sim_df = pd.DataFrame({sim: sim_vector}, index=investor_array_names)
    
    similarity_df = similarity_df.join(new_sim_df)
    
    #if ind == 1:
     #   similarity_df = similarity_df.sort_values(by=similarity[0], ascending=False).head(n)
     #   investors_top = list(similarity_df.index)
     #   investor_array = [invested_list[list(investors).index(x)] for x in investors_top]
    #    investors = investors_top.copy()
    #    ind += 1
    similarity_df = similarity_df.sort_values(by=similarity[0], ascending=False).head(n)

  return similarity_df

n_nearest_neighbours("Founders Fund", similarity = ("abs"), similarity_type = "o_to_o", n = len(master_organization_list_likes))
```
